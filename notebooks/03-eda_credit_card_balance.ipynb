Definição de porcentagem de nulos para cada coluna

```python
def porcentagem_nulos_df(df):
    return (
        df.isnull()
          .mean()
          .mul(100)
          .reset_index()
          .rename(columns={'index': 'coluna', 0: 'porcentagem_nulos'})
          .sort_values(by='porcentagem_nulos', ascending=True)
    )
```

```python
pn = porcentagem_nulos_df(dataset)
pn
```

- Desvio padrão para entender estabilidade dos limites dos cartões
- Mean para entender resumir os saldos dos cartões
- Colocar Last para entender o último saldo do cartão

```python
dataset_agg = dataset.groupby('SK_ID_CURR')['AMT_BALANCE'].agg(
    AMT_BALANCE_BIGGER = 'max',
    AMT_BALANCE_STD = 'std',
    AMT_BALANCE_MEAN = 'mean'
)
```

```python
last_balance = (
    dataset
    .sort_values(['SK_ID_CURR','MONTHS_BALANCE'])
    .groupby('SK_ID_CURR')
    .tail(1)[['SK_ID_CURR','AMT_BALANCE']]
    .rename(columns={'AMT_BALANCE':'AMT_BALANCE_LAST'})
)

last_balance
```

```python
last_balance = last_balance.set_index('SK_ID_CURR')

dataset_agg = dataset_agg.join(last_balance)
```
